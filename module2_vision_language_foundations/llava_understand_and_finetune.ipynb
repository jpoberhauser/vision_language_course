Models like LLaVA are used as bases for more interesting VQA like [this project](https://arxiv.org/pdf/2503.18712) that uses it as a base for action-recognition in videos

github repo is [here](https://github.com/LLaVA-Annonymous/LLaVA)

Lets review some things:

* LLaVA: Uses a CLIP Vit -> linear projection -> Vicuna/LLaMA. Projection is a linear layer of small MLPs.
* PaliGemma: USes a SigLip Vit -> linear projection -> Gemma. 

To train:
* LLaVA: 2 stage. 
  * 1. Freeze vision encoder + LLM and train only projection layer in image-caption pairs. 
  * 2. Unfreeze the LLM and fine-tune on visual instruction data
* PaliGemma: 3 stage.
  * 1. unimodal pretraining of both encoders separately
  * 2. multi-modal pretraining on image-text pairs on both vision and language components
  * 3. task-specific fine-tuning
